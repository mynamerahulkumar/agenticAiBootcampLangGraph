 langchain_huggingface library tool that allows us to use pre-trained models from Hugging Face to create embeddings. It then selects a specific model called "all-MiniLM-L6-v2" for this purpose and stores it in a variable called embeddings for later use.



```python
from langchain_huggingface import HuggingFaceEmbeddings
embeddings=HuggingFaceEmbeddings(model_name="all-MiniLM-L6-v2")
```

```python
text="this is hugginface embeddings"
query_result=embeddings.embed_query(text)
query_result
```

```python
len(query_result)
```

```python
doc_result = embeddings.embed_documents([text, "This is not a test document."])
doc_result[0]
```
