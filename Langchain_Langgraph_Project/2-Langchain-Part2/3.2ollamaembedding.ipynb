# Ollama

Ollama supports embedding models, making it possible to build retrieval augmented generation (RAG) applications that combine text prompts with existing documents or other data.

```python
from langchain_community.embeddings import OllamaEmbeddings
```

```python
embedding=OllamaEmbeddings(model='llama3.2:1b')
```

```python
embedding
```

```python
query="I am learning Agentic AI"
embed_vector=embedding.embed_query(query)
```

```python
embed_vector
```

```python
len(embed_vector)
```

```python
#embeded document
r1=embedding.embed_documents(
    ["Agentic AI backbone is generative AI",
    "Llama model is developed by Meta"]
)
```
